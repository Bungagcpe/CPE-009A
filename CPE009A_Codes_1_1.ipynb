```python
#Supplementary Activity_3
x_1 = float(input("Enter the First x-coordinate: "))
y_1 = float(input("Enter the First y-coordinate: "))
x_2 = float(input("Enter the Second x-coordinate: "))
y_2 = float(input("Enter the Second y-coordinate: "))
Distance = ((x_2 - x_1)**2 + (y_2 - y_1)**2)**0.5
print("The Distance Between the Two Points is", Distance)
```
